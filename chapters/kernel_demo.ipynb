# Kernel demo
This notebook plots the memory kernel $(t-\tau)^{\alpha-1}$ on $\tau\in[0,t]$ for different $\alpha$.

```python
import numpy as np
import matplotlib.pyplot as plt

t = 1.0
taus = np.linspace(0, t, 400)
alphas = [0.2, 0.5, 0.9]
plt.figure(figsize=(6,4))
for a in alphas:
    kernel = (t - taus)**(a-1)
    plt.plot(taus, kernel, label=f'alpha={a}')
plt.xlabel('tau')
plt.ylabel('kernel (t-tau)^{alpha-1}')
plt.legend()
plt.title('Memory kernel for different alpha')
plt.show()
```
